# Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the DataSet

```python
tele_df=pd.read_csv("Telco_Customer_Churn_Data.csv")
tele_df
```

# Seperating the Output Variable

```python
y=tele_df["Churn"]
tele_df.drop(columns=['Churn'],inplace=True)
tele_df
```

# Listing out Categorical and Continuous Variables

```python
tele_df.dtypes
tele_df["TotalCharges"]=pd.to_numeric(tele_df["TotalCharges"],downcast='float',errors='coerce')
tele_df['SeniorCitizen']=tele_df['SeniorCitizen'].astype(object)
tele_df.dtypes
```

```python
m=tele_df['TotalCharges'].mean()
tele_df['TotalCharges']=tele_df['TotalCharges'].fillna(m)
tele_df
```

```python
categorical=[]
continuous=[]
for col in list(tele_df):
    if(tele_df[col].dtype=='object'):
        categorical.append(col)
    else:
        continuous.append(col)
print(continuous)
print(categorical)
```

# Drop unwanted Columns

```python
for col in categorical:
    l=list(tele_df[col])
    if(len(set(l))==1 or len(set(l))==len(l)):
        tele_df.drop(columns=[col],inplace=True)
        print("Yes")
```

```python
categorical=[]
continuous=[]
for col in list(tele_df):
    if(tele_df[col].dtype=='object'):
        categorical.append(col)
    else:
        continuous.append(col)
print(continuous)
print(categorical)
```

# Corelation Check

```python
from scipy.stats import chi2_contingency as chi
corr_categorical=[]
for col in categorical:
    con_table=pd.crosstab(tele_df[col],y)
    p=chi(con_table)[1]
    if(p<0.05):
        corr_categorical.append(col)
corr_categorical
```

```python
from scipy.stats import f_oneway
def anova(categorical,continuous):
    l=[]
    for i in set(categorical):
        arg=[]
        for j in range(len(categorical)):
            if(categorical[j]==i):
                arg.append(continuous[j])
        l.append(arg)
    return f_oneway(*l)
for col in continuous:
        print(col)
        print(anova(list(y),list(tele_df[col])))
corr_continuous=continuous
```

```python
preprocessed_df=pd.DataFrame(columns=[])
for i in corr_categorical:
    preprocessed_df[i]=tele_df[i]
for i in corr_continuous:
    preprocessed_df[i]=tele_df[i]
preprocessed_df
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split as ttsplit
X_train,X_test,y_train,y_test=ttsplit(preprocessed_df,y,test_size=0.2,random_state=17)
X_train
```

```python
con=[]
cat=[]
for col in list(preprocessed_df):
    if(preprocessed_df[col].dtype=='object'):
        cat.append(col)
    else:
        con.append(col)
```

# Scaling continuous variables

```python
from sklearn.preprocessing import StandardScaler 
for col in corr_continuous:
    scaler=StandardScaler()
    train_scaled=scaler.fit_transform(np.array(X_train[col]).reshape(-1,1))
    X_train.drop(columns=col,inplace=True)
    X_train[col]=train_scaled
    test_scaled=scaler.transform(np.array(X_test[col]).reshape(-1,1))
    X_test.drop(columns=col,inplace=True)
    X_test[col]=test_scaled
```

# Encoding categorical variables and extracting final test and train sets

```python
X_train=pd.concat([pd.get_dummies(X_train[cat]),X_train[con]],axis=1)
X_test=pd.concat([pd.get_dummies(X_test[cat]),X_test[con]],axis=1)
```

# Training the dataset on Logistic Regression

```python
from sklearn.linear_model import LogisticRegression,Lasso
lr=LogisticRegression()
lr.fit(X_train,y_train)
lr_pred=lr.predict(X_test)
lr_pred
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_auc_score,roc_curve
print(accuracy_score(lr_pred,y_test))
print(precision_score(lr_pred,y_test,pos_label='Yes'))
print(recall_score(lr_pred,y_test,pos_label='Yes'))
print(f1_score(lr_pred,y_test,pos_label='Yes'))
print(roc_auc_score(pd.get_dummies(lr_pred),pd.get_dummies(y_test)))
```
